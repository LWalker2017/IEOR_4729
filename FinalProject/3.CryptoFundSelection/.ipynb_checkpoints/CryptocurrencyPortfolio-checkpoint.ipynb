# Crypto Fund Competition and Selection
- Let's get started

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
import copy
%matplotlib inline
plt.style.use('ggplot')
```

```python
tick_data=pd.read_csv("data/crypto_2y.csv")
tick_data['time']=pd.to_datetime(tick_data['time'])
tick_data = tick_data.set_index('time')
tick_data.head()
```

```python
tick_data.plot(figsize=(12, 8))
plt.title('Cryptocurrency Price',size=20)
plt.legend(loc='best')
plt.show()
```

```python
# Select the test period from 2018-12-01 to 2019-12-31, the rest period for backtesting
test_data = tick_data['20181201':'20191231']
train_data = tick_data['20200101':]
```

# Cryptocurrency Allocation with Sharpe Ratio
- We will assume a risk-free rate of 0% and target return of 0% for the time being

```python
# Calculate the daily Time-Weighted Average Price(TWAP) of the Cryptocurrency
TWAP = test_data.groupby(by=test_data.index.date).mean()
TWAP.index = pd.to_datetime(TWAP.index)
returns = TWAP.pct_change()
returns.dropna(inplace=True)
stats = returns.describe()
Sharpe_Ratio = stats.loc['mean'] / stats.loc['std']
Sharpe_Ratio_Allocation = Sharpe_Ratio/sum(Sharpe_Ratio)
Sharpe_Ratio_Allocation
```

```python
# Visualization
plt.style.use('seaborn-darkgrid')

opt_weights_all = Sharpe_Ratio_Allocation.sort_values(ascending=True)
colors = ['k','y','c','m','r','g','b']
opt_weights_all.plot(kind='barh', color=colors, figsize=(10,10), align='center', edgecolor="black", alpha=0.8)
plt.xlabel('Cryptocurrency weights')
plt.title('Cryptocurrency Allocation with Sharpe Ratio')
# plt.savefig('./Cryptocurrency_Allocation_with_Sharpe_Ratio.png')
plt.show()
```

# Cryptocurrency Allocation with Calmar Ratio

```python
geometric_monthly_return = (TWAP.loc['20191231']/TWAP.loc['20181201'])**(1/13)-1
```

```python
def MaxDrawdown(trading_days, return_list, label='BTC'):
    '''return_list is a 1-Dimension list or np.array'''
    i = np.argmax((np.maximum.accumulate(return_list) - return_list) / np.maximum.accumulate(return_list))  # ending point
    if i == 0:
        return 0
    j = np.argmax(return_list[:i])  # starting point
    Max_Drawdown_Rate = (return_list[j] - return_list[i]) / (return_list[j])  # mdd
    plt.plot(trading_days, return_list, label=label)
    plt.title("Max Drawdown", color='k', size=15)
    plt.xlabel("Date", size=10)
    plt.ylabel("Net Worth", size=10)
    plt.plot([trading_days[i], trading_days[j]], [return_list[i], return_list[j]], 'o', color="r", markersize=10)
    plt.legend()
    plt.grid(True)
    plt.show()
    plt.close()
    return Max_Drawdown_Rate

MDD = []
for col in TWAP.columns:
    MDD.append(MaxDrawdown(TWAP.index.values, TWAP[col].values, col))
```

```python
Calmar_Ratio = geometric_monthly_return/np.array(MDD)
Calmar_Ratio_Allocation = Calmar_Ratio/sum(Calmar_Ratio)
Calmar_Ratio_Allocation
```

```python
# Visualization
plt.style.use('seaborn-darkgrid')

opt_weights_all = Calmar_Ratio_Allocation.sort_values(ascending=True)
colors = ['k','y','c','m','r','g','b']
opt_weights_all.plot(kind='barh', color=colors, figsize=(10,10), align='center', edgecolor="black", alpha=0.8)
plt.xlabel('Cryptocurrency weights')
plt.title('Cryptocurrency Allocation with Calmar Ratio')
# plt.savefig('./Cryptocurrency_Allocation_with_Calmar_Ratio.png')
plt.show()
```

# Cryptocurrency Allocation with Minimum Variance Model

```python
# calculate the Covariance of the rate of returns of Cryptocurrency
# covariance_of_returns = np.cov(returns.values, rowvar=False)
# matrix_inverse = np.linalg.inv(covariance_of_returns)

covariance_of_returns = np.corrcoef(returns.values, rowvar=False)
e = np.ones((7,1))
Markowitz = np.linalg.solve(covariance_of_returns, e)
Markowitz = Markowitz/np.sum(Markowitz)
Markowitz_Allocation = pd.Series(data=Markowitz.reshape(-1), index=Sharpe_Ratio_Allocation.index)
Markowitz_Allocation
```

```python
# Visualization
plt.style.use('seaborn-darkgrid')

opt_weights_all = Markowitz_Allocation.sort_values(ascending=True)
colors = ['k','y','c','m','r','g','b']
opt_weights_all.plot(kind='barh', color=colors, figsize=(10,10), align='center', edgecolor="black", alpha=0.8)
plt.xlabel('Cryptocurrency weights')
plt.title('Cryptocurrency Allocation with Markowitz Model')
# plt.savefig('./Cryptocurrency_Allocation_with_Markowitz_Model.png')
plt.show()
```

# Cryptocurrency Allocation with Sortino Ratio
- For the Sortino Ratio, we calculate the downside deviation of the expected returns by taking the difference between each period’s return and the target return. If a period’s return is greater than the target return, the difference is simply set to 0. Then, we square the value of the difference. Next, we calculate the average of all squared differences. The square root of the average is the downside deviation.
- Sortino Ratio = (r – rf ) / σd
- where: r = expected return of strategy x; rf = risk-free rate of return; σd = downside deviation of expected return of strategy x

```python
# rfr = 0
# target = 0
# returns = df['Returns']
# sharpe_ratio = ((returns.mean() - rfr) / returns.std())
# print(sharpe_ratio)
# df['downside_returns'] = 0
# df.loc[df['Returns'] < target, 'downside_returns'] = df['Returns']**2
# expected_return = df['Returns'].mean()
# down_stdev = np.sqrt(df['downside_returns'].mean())
# sortino_ratio = (expected_return - rfr)/down_stdev
# print(sortino_ratio)

downside_returns = returns.copy()
downside_returns.iloc[returns.values>=0] = 0
downside_returns = downside_returns**2

avg_return = returns.mean(axis=0)
down_stdev = np.sqrt(downside_returns.mean(axis=0))
Sortino_Ratio = avg_return/down_stdev
Sortino_Ratio_Allocation = Sortino_Ratio/sum(Sortino_Ratio)
Sortino_Ratio_Allocation
```

```python
# Visualization
plt.style.use('seaborn-darkgrid')

opt_weights_all = Sortino_Ratio_Allocation.sort_values(ascending=True)
colors = ['k','y','c','m','r','g','b']
opt_weights_all.plot(kind='barh', color=colors, figsize=(10,10), align='center', edgecolor="black", alpha=0.8)
plt.xlabel('Cryptocurrency weights')
plt.title('Cryptocurrency Allocation with Sortino Ratio')
# plt.savefig('./Cryptocurrency_Allocation_with_Sortino_Ratio.png')
plt.show()
```

# Cryptocurrency Allocation with Equal Weights

```python
Equal_Weight_Allocation = pd.Series(data=np.ones(7)/7, index=Sortino_Ratio_Allocation.index)
Equal_Weight_Allocation
```

```python
# Visualization
plt.style.use('seaborn-darkgrid')

opt_weights_all = Equal_Weight_Allocation.sort_values(ascending=True)
colors = ['k','y','c','m','r','g','b']
opt_weights_all.plot(kind='barh', color=colors, figsize=(10,10), align='center', edgecolor="black", alpha=0.8)
plt.xlabel('Cryptocurrency weights')
plt.title('Cryptocurrency Allocation with Equal Weights')
# plt.savefig('./Cryptocurrency_Allocation_with_Equal_Weights.png')
plt.show()
```

# Crypto-Asset Allocation Visualization
- Sharpe Ratio, Calmar Ratio, Sortino Ratio, Markowitz Model, Equal Weitght
- Which crypto allocation would give the best P&L?

```python
plt.subplots(figsize=(12,8))

plt.plot(Markowitz_Allocation.index, Markowitz_Allocation.values, label='Markowitz')
plt.plot(Sharpe_Ratio_Allocation.index, Sharpe_Ratio_Allocation.values, label='Sharpe')
plt.plot(Calmar_Ratio_Allocation.index, Calmar_Ratio_Allocation.values, label='Calmar')
plt.plot(Sortino_Ratio_Allocation.index, Sortino_Ratio_Allocation.values, label='Sortino')
plt.plot(Equal_Weight_Allocation.index, Equal_Weight_Allocation.values, label='Equal')

plt.scatter(Markowitz_Allocation.index, Markowitz_Allocation.values, linewidths=3)
plt.scatter(Sharpe_Ratio_Allocation.index, Sharpe_Ratio_Allocation.values, linewidths=3)
plt.scatter(Calmar_Ratio_Allocation.index, Calmar_Ratio_Allocation.values, linewidths=3)
plt.scatter(Sortino_Ratio_Allocation.index, Sortino_Ratio_Allocation.values, linewidths=3)
plt.scatter(Equal_Weight_Allocation.index, Equal_Weight_Allocation.values, linewidths=3)

plt.axhline(y=0, linestyle="--", c="green") # add horizontal line
plt.xlabel('Cryptocurrency')
plt.ylabel('Portfolio Weights')
plt.title('Crypto-Asset Allocation', size=20)
plt.legend(loc='best')
plt.grid(True)
# plt.savefig('./Crypto_Asset_Allocation.png')
plt.show()
```

```python
Crypto_Allocation = pd.concat([Markowitz_Allocation, Sharpe_Ratio_Allocation, Calmar_Ratio_Allocation,\
                               Sortino_Ratio_Allocation, Equal_Weight_Allocation], axis=1)
Crypto_Allocation.columns = ['Markowitz', 'Sharpe', 'Calmar', 'Sortino', 'Equal']
Crypto_Allocation
```

# Backtesting

```python
# Select the backtesting period from 2020-01-01 to 2020-04-30
train_data
```

```python
capital = 1000000
transation_cost = 0.001

Allocation_Array = Crypto_Allocation.values * capital
Bid_Price = train_data.iloc[0, :].values.reshape((7,1))
Bid_Price = Bid_Price * (1+transation_cost)
Initial_Shares = Allocation_Array / Bid_Price

Portfolio_Balance = pd.DataFrame(data=0, index=train_data.index, columns=Crypto_Allocation.columns, dtype=np.float64)

for date in train_data.index:
    Bid_Price = train_data.loc[date, :].values.reshape((7,1))
    Portfolio_Balance.loc[date] = np.sum(Initial_Shares*Bid_Price, axis=0)
    
Portfolio_Balance
```

```python
Portfolio_Balance.plot(figsize=(18, 9))
plt.title('Portfolio Performance Graph', size=20)
plt.ylabel('Net Worth')
plt.legend(loc='best', fontsize='xx-large')
plt.show()
```

```python
Performance_Evaluation = Portfolio_Balance.pct_change().describe()
Performance_Evaluation
```

```python
Portfolio_Sharpe_Ratio = Performance_Evaluation.loc['mean'] / Performance_Evaluation.loc['std']
Portfolio_Sharpe_Ratio
```

```python
plt.subplots(figsize=(12,8))
plt.bar(Portfolio_Sharpe_Ratio.index, Portfolio_Sharpe_Ratio.values, color='bgrym')
plt.xlabel('Crypto Allocation', fontsize=20)
plt.ylabel('Portfolio Sharpe Ratio')
plt.title('Sharpe Ratio Comparision', size=20)
plt.grid(True)
# plt.text to show the numeric value of each bar
for a,b in Portfolio_Sharpe_Ratio.items():  
     plt.text(a, b, '%f' % b, ha='center', va= 'bottom', fontsize=15)  
# plt.savefig('./Portfolio_Sharpe_Ratio_Comparision.png')
plt.show()
```
