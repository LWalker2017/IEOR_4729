```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
%matplotlib inline
```

```python
tick_data=pd.read_csv("data/crypto_2y.csv")
tick_data['time']=pd.to_datetime(tick_data['time'])
tick_data = tick_data.set_index('time')
tick_data.head()
```

```python
color = sns.color_palette()
sns.set_style('whitegrid')

#correlation map figure
def correlation_map(data):
    correlation = data.corr()
    plt.subplots(figsize=(10,8))
    plt.title('Correlation Map', fontsize=20)
    sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns,cmap="RdBu_r")
    # plt.savefig('Correlation_Map.png')
    
correlation_map(tick_data)
```

```python
plt.style.use('ggplot')
# ETH
ETH = tick_data['ETH'].pct_change()
ETH.dropna(inplace=True)
ETH_mean = np.mean(ETH)
ETH_sigma = np.std(ETH,ddof=1)
ETH_x = np.linspace(-0.07, 0.07, 2000)
ETH_y = 1/np.sqrt(2*np.pi)/ETH_sigma * np.exp(-(ETH_x-ETH_mean)**2/2/ETH_sigma**2)
# BTC
BTC = tick_data['BTC'].pct_change()
BTC.dropna(inplace=True)
BTC_mean = np.mean(BTC)
BTC_sigma = np.std(BTC,ddof=1)
BTC_x = np.linspace(-0.07, 0.07, 2000)
BTC_y = 1/np.sqrt(2*np.pi)/BTC_sigma * np.exp(-(BTC_x-BTC_mean)**2/2/BTC_sigma**2)
# XRP
XRP = tick_data['XRP'].pct_change()
XRP.dropna(inplace=True)
XRP_mean = np.mean(XRP)
XRP_sigma = np.std(XRP,ddof=1)
XRP_x = np.linspace(-0.07, 0.07, 2000)
XRP_y = 1/np.sqrt(2*np.pi)/XRP_sigma * np.exp(-(XRP_x-XRP_mean)**2/2/XRP_sigma**2)
# BCH
BCH = tick_data['BCH'].pct_change()
BCH.dropna(inplace=True)
BCH_mean = np.mean(BCH)
BCH_sigma = np.std(BCH,ddof=1)
BCH_x = np.linspace(-0.07, 0.07, 2000)
BCH_y = 1/np.sqrt(2*np.pi)/BCH_sigma * np.exp(-(BCH_x-BCH_mean)**2/2/BCH_sigma**2)
# USDT
USDT = tick_data['USDT'].pct_change()
USDT.dropna(inplace=True)
USDT_mean = np.mean(USDT)
USDT_sigma = np.std(USDT,ddof=1)
USDT_x = np.linspace(-0.07, 0.07, 2000)
USDT_y = 1/np.sqrt(2*np.pi)/USDT_sigma * np.exp(-(USDT_x-USDT_mean)**2/2/USDT_sigma**2)
# LTC
LTC = tick_data['LTC'].pct_change()
LTC.dropna(inplace=True)
LTC_mean = np.mean(LTC)
LTC_sigma = np.std(LTC,ddof=1)
LTC_x = np.linspace(-0.07, 0.07, 2000)
LTC_y = 1/np.sqrt(2*np.pi)/LTC_sigma * np.exp(-(LTC_x-LTC_mean)**2/2/LTC_sigma**2)
# EOS
EOS = tick_data['EOS'].pct_change()
EOS.dropna(inplace=True)
EOS_mean = np.mean(EOS)
EOS_sigma = np.std(EOS,ddof=1)
EOS_x = np.linspace(-0.07, 0.07, 2000)
EOS_y = 1/np.sqrt(2*np.pi)/EOS_sigma * np.exp(-(EOS_x-EOS_mean)**2/2/EOS_sigma**2)

plt.subplots(figsize=(12,8))
plt.plot(ETH_x, ETH_y, label='ETH', linewidth=5)
plt.plot(BTC_x, BTC_y, label='BTC', linewidth=5)
plt.plot(XRP_x, XRP_y, label='XRP', linewidth=5)
plt.plot(USDT_x, USDT_y, label='USDT', linewidth=2)
plt.plot(BCH_x, BCH_y, label='BCH', linewidth=5)
plt.plot(LTC_x, LTC_y, label='LTC', linewidth=5)
plt.plot(EOS_x, EOS_y, label='EOS', linewidth=5)
plt.xlabel('Asset return')
plt.ylabel('Probability')
plt.title('Asset Return Distribution', fontsize=15)
plt.grid(True)
plt.legend()
# plt.savefig('Asset_Return_Distribution.png')
plt.show()
```
