```python
# Load NIKE Data
taq = pd.read_csv( '../data/NKE_TAQ_Mar20.gz', index_col = 0, parse_dates = True )
# filter on the date
test_day = '2020-03-25'
day = taq[test_day]
day
```

## VWAP Target Function

```python
# the result of vwap_coefs comes from Question 3
vwap_coefs = [ 6.40040086e-03, -4.74660554e-05,  2.67999184e-07, -7.51363856e-10, 7.97204428e-13]
```

```python
order_quantity = 300000
# order_side = 'b'
order_side = 's'
results = vw.algo_loop(day, order_side, order_quantity, vwap_coefs)
```
